# technews parser to mongodb

```python
import re
import pymongo
import datetime
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pymongo import MongoClient
from dateutil.relativedelta import relativedelta
%matplotlib inline
```

### Url set, then use bs4 to get page

```python
start_url = "https://technews.tw/?s=%E6%AF%94%E7%89%B9%E5%B9%A3"
```

```python
title_list = []
link_list = []
date_list = []
source_list = []
content_list = []
```

### Scrape_Parser Functions

```python
def parse(url):
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, "html.parser")
    print('##################################parse...', url)
    for article in soup.select("article"):
        link = article.a['href']
        parse_detail(link)
    try:
        pre_page = soup.select('.nav-previous a')[0].attrs['href']  # 向前換頁
        if int(pre_page.split('/')[-2]) < 15:
            print('##################################change page to', pre_page)
            parse(pre_page)
    except:
        print("no page")
```

```python
def parse_detail(url):
    print('parse_detail...', url)
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, "html.parser")
    
    predate = soup.select(".content span[class=\"body\"]")[1].text #'2017 年 07 月 17 日 13:11 '
    year = int(predate.split('年')[0].strip())
    month = int(predate.split('月')[0].split('年')[1].strip())
    day = int(predate.split('月')[1].split('日')[0].strip())
    hour = int(predate.split('日')[1].strip().split(':')[0])
    minute = int(predate.split('日')[1].strip().split(':')[1])
    sfttime = datetime.datetime(year, month, day, hour, minute) #'2017-07-17 13:11:00'

    title = soup.select('.entry-title a')[0].text
    link = soup.select('.entry-title a')[0].attrs['href']
    date = str(sfttime)
    content = soup.select(".indent")[0].text
    source = 'technews'
    
    title_list.append(title)
    link_list.append(link)
    source_list.append(source)
    date_list.append(date)
    content_list.append(content)
    print('parse_detail...Completed', url)
```

### Catch data

```python
parse(start_url)
```

### to pd.dataframe

```python
# show data
all_dict = {'Title':title_list,
        'Source':source_list,
        'Link':link_list,
        'Date':date_list,
        'Content':content_list
       }
```

```python
df = pd.DataFrame(all_dict) 
df
```

### 整理時間格式

```python
for index, row in df.iterrows():
    try:
        x = datetime.datetime.strptime(row['Date'], '%Y-%m-%d %H:%M:%S')
        df.set_value(index, 'Date', x)
    except:
        print(index, 'error')
```

```python
df.sort_values(by='Date')
```

### list，用於存入db

```python
tt = []
for index, row in df.iterrows():
    temp = {
        "Date": row['Date'],
        "Link": row['Link'],
        "Source": row['Source'],
        "Title": row['Title'],
        "Content": row['Content']
    }
    tt.append(temp)
```

```python
tt
```

### to mongodb

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.bitcoinnews_db
for item in tt:
    db.newsCollection.insert_one(item)
```
